# Globally averaged variables in FOCI-OpenIFS

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cmocean 
```

```python
import foci_functions
```

```python
exp_list = ['FOCI_GJK029']
time_list = [slice('2050-01-01','2550-01-01')]
```

## Read OpenIFS data

```python
ds_oifs_1y = foci_functions.read_openifs(exp_list, time_list, grid='regular_sfc', freq='1y')
```

```python
ds_oifs_1y[0]
```

## Read NEMO data

```python
ds_nemo_1y = foci_functions.read_nemo(exp_list, time_list, grid='grid_T', freq='1y')
```

```python
ds_nemo_1y[0]
```

## Read CERES obs

```python
ds_ceres = xr.open_dataset('data/CERES_timmean.nc')
```

```python
## Plot time series
```

```python
fig1, ax1 = plt.subplots(2,3,figsize=(10,6))
axs = ax1.flat

runmean = 10

for (ds,ds_o) in zip(ds_oifs_1y, ds_nemo_1y):
    
    # TOA SW
    tsr = foci_functions.global_mean(ds['tsr']) / (6*3600.)
    tsr.rolling(time=runmean,center=True).mean().plot(ax=axs[0])
    # In CERES, solar_mon is incoming, and toa_sw_all_mon is reflected
    # In OpenIFS, tsr is incoming minus reflected
    tsr_o = foci_functions.global_mean(ds_ceres['solar_mon'] - ds_ceres['toa_sw_all_mon'])
    axs[0].axhline(y=tsr_o.data, linestyle='--', color='k')
    print(tsr.mean('time').load().data, tsr_o.data)
    
    # TOA LW
    ttr = foci_functions.global_mean(ds['ttr']) / (6*3600.)
    ttr.rolling(time=runmean,center=True).mean().plot(ax=axs[1])
    ttr_o = foci_functions.global_mean(-ds_ceres['toa_lw_all_mon'])
    axs[1].axhline(y=ttr_o.data, linestyle='--', color='k')
    print(ttr.mean('time').load().data, ttr_o.data)
    
    # TOA net
    p1, = (ttr+tsr).rolling(time=runmean,center=True).mean().plot(ax=axs[2],label='TOA')
    axs[2].axhline(y=(ttr_o.data+tsr_o.data), linestyle='--', color=p1.get_color())
    print((ttr+tsr).mean('time').load().data, (ttr_o+tsr_o).data)
    
    # Ocean surface net
    sflx = foci_functions.global_mean_nemo(ds_o['sohefldo'], ds_o['tmask'][0,:,:], ds_o['tarea'])
    p2, = sflx.rolling(time=runmean,center=True).mean().plot(ax=axs[2], label='Ocean')
    print(sflx.mean('time').load().data)
    
    # Planetary albedo
    # tisr is incoming solar, tsr is net solar. Both at TOA. 
    # Reflected solar at TOA is thus tsru = tsr - tisr. 
    # Albedo is tsru / tisr 
    tisr = foci_functions.global_mean(ds['tisr']) / (6*3600.)
    tisr_o = foci_functions.global_mean(ds_ceres['solar_mon'])
    alb = np.abs(tsr - tisr) / tisr.where(tisr != 0) 
    alb_o = np.abs(tsr_o - tisr_o) / tisr_o.where(tisr_o != 0)
    alb.rolling(time=runmean,center=True).mean().plot(ax=axs[3])
    axs[3].axhline(alb_o.data, linestyle='--', c='k')
    print(alb.mean('time').load(), alb_o.load().data)
    
    # TOA CRE
    # We take TOA clear-sky minus TOA
    # Clear-sky is computed by radiation scheme assuming no clouds
    toa_c = foci_functions.global_mean(ds['tsrc'] + ds['ttrc']) / (6*3600.)
    p1,=((ttr+tsr)-toa_c).rolling(time=runmean,center=True).mean().plot(ax=axs[4],label='TOA')
    toa_cre = foci_functions.global_mean((ds_ceres['toa_cre_sw_mon'] - ds_ceres['toa_cre_lw_mon']))
    axs[4].axhline(y=(toa_cre).data, linestyle='--', color=p1.get_color())
    print(((ttr+tsr)-toa_c).mean('time').load().data, toa_cre.data)
    
    # Surface CRE
    # We take surface clear-sky minus surface
    sfc   = foci_functions.global_mean(ds['ssr']  + ds['str']) / (6*3600.)
    sfc_c = foci_functions.global_mean(ds['ssrc'] + ds['strc']) / (6*3600.)
    p2,=(sfc-sfc_c).rolling(time=runmean,center=True).mean().plot(ax=axs[4],linestyle='-',label='Surface')
    sfc_cre = foci_functions.global_mean((ds_ceres['sfc_cre_net_sw_mon'] - ds_ceres['sfc_cre_net_lw_mon']))
    axs[4].axhline(y=(sfc_cre).data, linestyle='--', color=p2.get_color())
    print((sfc-sfc_c).mean('time').load().data, sfc_cre.data)
    
    # Cloud cover
    cc = foci_functions.global_mean(ds['tcc'])
    cc_o = foci_functions.global_mean(ds_ceres['cldarea_total_daynight_mon']) * 0.01
    cc.rolling(time=runmean,center=True).mean().plot(ax=axs[5])
    axs[5].axhline(y=cc_o.data, linestyle='--', color='k')
    print(cc.mean('time').load().data, cc_o.data)
    
axs[0].set_ylim([238,242])
axs[0].set_title('a) TOA net SW flux')
axs[0].set_ylabel('[W/m2]')

axs[1].set_ylim([-242,-238])
axs[1].set_title('b) TOA net LW flux')
axs[1].set_ylabel('[W/m2]')

axs[2].set_ylim([-1,1])
axs[2].set_title('c) Net heat flux')
axs[2].set_ylabel('[W/m2]')
axs[2].legend()

axs[3].set_ylim([0.285, 0.3])
axs[3].set_title('d) Planetary albedo')
axs[3].set_ylabel('')

axs[4].set_ylim([-80,-15])
axs[4].set_title('e) Cloud radiative forcing')
axs[4].set_ylabel('[W/m2]')
axs[4].legend()

axs[5].set_ylim([0.62,0.7])
axs[5].set_title('f) Cloud cover')
axs[5].set_ylabel('[fraction]')

#axs[5].axis('off')

for ax in axs:
    ax.set_xlabel('Model year')

fig1.tight_layout()
fig1.savefig('plots/global_mean_radiation_budget.pdf',format='pdf')
```

```python
fig1, ax1 = plt.subplots(2,3,figsize=(12,7))
axs = ax1.flat

runmean = 10

for (ds,ds_o) in zip(ds_oifs_1y, ds_nemo_1y):
    
    # TOA SW
    tsr = ds['tsr'].mean('lon').mean('time') / (6*3600.)
    tsr.plot(ax=axs[0])
    # Obs
    tsr_o = (ds_ceres['solar_mon'] - ds_ceres['toa_sw_all_mon']).mean('lon')
    tsr_o.plot(ax=axs[0], linestyle='--', color='k')
    
    # TOA LW
    ttr = ds['ttr'].mean('lon').mean('time') / (6*3600.)
    ttr.plot(ax=axs[1])
    ttr_o = (-ds_ceres['toa_lw_all_mon']).mean('lon')
    ttr_o.plot(ax=axs[1], linestyle='--', color='k')
    
    # TOA net
    p1, = (ttr+tsr).plot(ax=axs[2],label='TOA')
    (ttr_o+tsr_o).plot(ax=axs[2], linestyle='--', color=p1.get_color())
    
    # Ocean surface net
    sflx, lat = foci_functions.zonal_mean_nemo(ds_o['sohefldo'].mean('time'), ds_o['tmask'][0,:,:], ds_o['tarea'])
    p2, = axs[2].plot(lat, sflx, label='Ocean')
    
    # Planetary albedo
    # tisr is incoming solar, tsr is net solar. Both at TOA. 
    # Reflected solar at TOA is thus tsru = tsr - tisr. 
    # Albedo is tsru / tisr 
    tisr = ds['tisr'].mean('lon').mean('time') / (6*3600.)
    tisr_o = ds_ceres['solar_mon'].mean('lon')
    alb = np.abs(tsr - tisr) / tisr.where(tisr != 0) 
    alb_o = np.abs(tsr_o - tisr_o) / tisr_o.where(tisr_o != 0)
    alb.plot(ax=axs[3])
    alb_o.plot(ax=axs[3], linestyle='--', c='k')
    
    # TOA CRE
    # We take TOA clear-sky minus TOA
    # Clear-sky is computed by radiation scheme assuming no clouds
    toa_c = (ds['tsrc'] + ds['ttrc']).mean('lon').mean('time') / (6*3600.)
    p1,=((ttr+tsr)-toa_c).plot(ax=axs[4],label='TOA')
    toa_cre = (ds_ceres['toa_cre_sw_mon'] - ds_ceres['toa_cre_lw_mon']).mean('lon')
    toa_cre.plot(ax=axs[4], linestyle='--', color=p1.get_color())
    
    # Surface CRE
    # We take surface clear-sky minus surface
    sfc   = (ds['ssr']  + ds['str']).mean('lon').mean('time') / (6*3600.)
    sfc_c = (ds['ssrc'] + ds['strc']).mean('lon').mean('time') / (6.*3600)
    p2,=(sfc-sfc_c).plot(ax=axs[4],linestyle='-',label='Surface')
    sfc_cre = (ds_ceres['sfc_cre_net_sw_mon'] - ds_ceres['sfc_cre_net_lw_mon']).mean('lon').mean('time')
    (sfc_cre).plot(ax=axs[4], linestyle='--', color=p2.get_color())
    
    # Cloud cover
    cc = ds['tcc'].mean('lon').mean('time')
    cc_o = (ds_ceres['cldarea_total_daynight_mon']).mean('lon') * 0.01
    cc.plot(ax=axs[5])
    cc_o.plot(ax=axs[5], linestyle='--', color='k')
    
#axs[0].set_ylim([238,242])
axs[0].set_title('a) TOA net SW flux')
axs[0].set_ylabel('[W/m2]')

#axs[1].set_ylim([-242,-238])
axs[1].set_title('b) TOA net LW flux')
axs[1].set_ylabel('[W/m2]')

#axs[2].set_ylim([-1,1])
axs[2].set_title('c) Net heat flux')
axs[2].set_ylabel('[W/m2]')
axs[2].legend()

axs[3].set_title('d) Planetary albedo')
axs[3].set_ylabel('')

#axs[3].set_ylim([-80,-15])
axs[4].set_title('e) Cloud radiative forcing')
axs[4].set_ylabel('[W/m2]')
axs[4].legend()

#axs[4].set_ylim([0.5,0.75])
axs[5].set_title('f) Cloud cover')
axs[5].set_ylabel('[fraction]')

for ax in axs:
    ax.set_xlabel('Latitude [degN]')

fig1.tight_layout()
fig1.savefig('plots/global_mean_radiation_budget_zonmean.pdf',format='pdf')
```
